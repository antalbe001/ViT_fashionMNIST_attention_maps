```python
from PIL import Image
from pathlib import Path
from IPython.display import display


# Apri i file immagine
# Ensure the paths are correctly constructed
image_dir = Path('./IMAGES')
grafico1 = Image.open(image_dir / 'orig.png')
grafico2 = Image.open(image_dir / 'mean.png')
grafico3 = Image.open(image_dir / 'min.png')
grafico4 = Image.open(image_dir / 'max.png')
grafico5 = Image.open(image_dir / 'rollout.png')

# Trova la larghezza e altezza massima
altezza = min(grafico1.height, grafico2.height)
larghezza = grafico1.width + grafico2.width + grafico3.width + grafico4.width + grafico5.width

# Crea una nuova immagine vuota con la larghezza e altezza totale
immagine_comb = Image.new('RGB', (larghezza, altezza))

# Incolla i grafici uno sotto l'altro
immagine_comb.paste(grafico1, (0, 0))
immagine_comb.paste(grafico2, (grafico1.width, 0))
immagine_comb.paste(grafico3, (grafico1.width + grafico2.width, 0))
immagine_comb.paste(grafico4, (grafico1.width + grafico2.width + grafico3.width, 0))
immagine_comb.paste(grafico5, (grafico1.width + grafico2.width + grafico3.width + grafico4.width, 0))

immagine_comb.save('./IMAGES/FINALS/grafico1.png')

display(immagine_comb)
```
